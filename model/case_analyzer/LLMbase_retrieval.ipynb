```python
!pip install -U -q sentence-transformers
from sentence_transformers import SentenceTransformer
import numpy as np
from tqdm import tqdm
from scipy.spatial.distance import cosine
```

```python
model = SentenceTransformer('ahdsoft/persian-sentence-transformer-news-wiki-pairs-v3')
```

```python
with open('case.txt', 'r', encoding = "utf-8") as file:
        case_text = file.read()
case_list = case_text.split("<#>")
case_list = case_list[:500]
```

```python
def embeding_list (list):
  result = []
  for text in tqdm(list):
    embedding = model.encode(text)
    result.append(embedding)
  return(result)
```

```python
len(case_list)
```

```python
emb = embeding_list(case_list)
np.savez('case.np', *emb)
```

```python
def cosine_similarity(a, b):
    dot_product = np.dot(a, b)
    norm_a = np.linalg.norm(a)
    norm_b = np.linalg.norm(b)
    similarity = dot_product / (norm_a * norm_b)
    return similarity

# Example arrays
array1 = emb[1]
array2 = emb[2]

# Calculate cosine similarity
similarity = cosine_similarity(array1, array2)

print("Cosine Similarity:", similarity)
```

```python
similarity = np.zeros((len(emb), len(emb)))
for i in tqdm(range(len(emb))):
  for j in range(len(emb)):
    similarity[i,j] = 1 - cosine(emb[i], emb[j])
```

```python
sorted_indices = np.argsort(-similarity[0,:])
sorted_indices[:11]
```

```python
np.save('case_similarity.npy', similarity)
```
